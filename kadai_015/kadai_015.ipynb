```python
#クラスを定義する
class Human:
  def __init__(self):
    #属性を定義する
    self.name=""
    self.age=""

  #メソッドを定義する
  def printinfo(self):
    print(self.name)
    print(self.age)

#インスタンスを変数に代入する
human = Human()

#メソッドにアクセスして実行する
human.name="花子"
human.age=20

# 属性にアクセスし、値を出力する
print(human.name)
print(human.age)
```

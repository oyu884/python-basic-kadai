```python
#ランダムな整数を利用するためにrandomモジュールをインポート
import random

#変数varに3～30までのランダムな整数を代入する
var=random.randint(3,30)

#変数varの値を出力
print(var)

#変数varが3の倍数の場合処理を行う
if var%3==0:
  print("Fizz")

#変数varが5の倍数の場合処理を行う
elif var%5==0:
  print("Buzz")

#すべての条件が成り立つ場合にのみ処理を行う（変数varの値が3の倍数と5の倍数の両方に該当する場合）
if var%3==0 and var%5==0:
  print("FizzBuzz")

else:
  print(var)
```

```python
#クラスを定義する
class Human:
  def __init__(self,name,age):
    #属性を定義する
    self.name=name
    self.age=age

  #メソッドを定義する(大人かどうかを判定する)
  def check_adult(self):
    if self.age >=20:
        print(f"{self.name}は大人です")
    if self.age<20:
        print(f"{self.name}は大人ではありません")

#人物情報（ディクショナリ）
human_data={"花子":19,"太郎":22,"一郎":35,"二郎":16}

#Humanインスタンスを格納するリスト
humans=[]

#繰り返し処理でインスタンスを生成しリストに追加
for name, age in human_data.items():
    humans.append(Human(name, age))

#リストの要素数分だけメソッドを呼び出す
for human in humans:
    human.check_adult()
```
